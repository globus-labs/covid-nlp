## RDKit: Simple Visualization of Molecules in Argonne COVID19 Molecule Dataset

```python
import pandas as pd
from rdkit.Chem.Draw import IPythonConsole
from rdkit import Chem
```

```python
df = pd.read_csv('./data/literature_molecules.csv')
df.head(4)
```

```python
n_mols = 8
ms = [Chem.MolFromSmiles(sm) for sm in df['smiles'][0:n_mols]]
Chem.Draw.MolsToGridImage(ms,molsPerRow=3, subImgSize=(250, 250), legends=["{}".format(x['molecule']) for i, x in df[0:n_mols].iterrows()])
```
